```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
df=pd.read_csv("Camp_Market.csv", sep=";")
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], utc=False)
df.info()
```

```python
df.describe()
```

```python
#Vérifier si des valeurs sont nulles

df.isnull().sum()
```

```python
df=df.drop(columns=["Z_CostContact", "Z_Revenue"],axis=1)
df.head()
```

```python
#Vérifier quelles donénes sont manquantes

sns.heatmap(df.isnull())
```

```python
#Prendre les revenus moyens pour compléter les données manquantes puis vérifier qu'aucune donnée n'est manquante maintenant

df['Income'] = df['Income'].fillna(df.groupby('Education')['Income'].transform('mean'))
sns.heatmap(df.isnull())
```

```python
df_numeric = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(20,20))
sns.heatmap(df_numeric.corr(), annot=True, cmap="YlGnBu",)
plt.show()
```

```python
corr = df_numeric.corr()
c1 = corr.abs().unstack()
c1.sort_values(ascending = False)[24:50:2]
```

```python
plt.figure()
df.boxplot(column=["Year_Birth"])
```

```python
df = df[df['Year_Birth'] > 1900]
df = df[df['Income'] < 200000]
numeric_cols = df.select_dtypes(include=['float64', 'int64'])
for col in numeric_cols.columns:
    plt.figure()
    numeric_cols.boxplot(column=[col])
    plt.title(f'Graphique à moustaches pour {col}')
    plt.ylabel(col)
    plt.show()
```

```python
current_year = datetime.now().year
df['Age'] = current_year - df['Year_Birth']
df = df.drop(columns=['Year_Birth'])
df.head()
```

```python
df.to_csv("DataOut.csv", index=False)
```
