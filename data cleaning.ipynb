```python
#VÃ©rifier si des valeurs sont nulles

df.isnull().sum()
```

```python
df=df.drop(columns=["Z_CostContact", "Z_Revenue"],axis=1)
df.head()
```

```python
#VÃ©rifier quelles donÃ©nes sont manquantes

sns.heatmap(df.isnull())
```

```python
#Prendre les revenus moyens pour complÃ©ter les donnÃ©es manquantes puis vÃ©rifier qu'aucune donnÃ©e n'est manquante maintenant

df['Income'] = df['Income'].fillna(df['Income'].mean())
sns.heatmap(df.isnull())
```

```python
df_numeric = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(20,20))
sns.heatmap(df_numeric.corr(), annot=True)
plt.show()
```

```python
corr = df_numeric.corr()
c1 = corr.abs().unstack()
c1.sort_values(ascending = False)[24:50:2]
```
